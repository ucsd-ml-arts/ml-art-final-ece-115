```python
#import necessary libraries
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import PIL
from PIL import Image
```

```python
#show normalized image
pic1 = plt.imread('./samples_content/fallen_star.jpg')/255  # dividing by 255 to bring the pixel values between 0 and 1
print(pic1.shape)
plt.imshow(pic1)
```

```python
basewidth = 300
img = Image.open('./samples_content/fallen_star.jpg')
wpercent = (basewidth / float(img.size[0]))
hsize = int((float(img.size[1]) * float(wpercent)))
img = img.resize((basewidth, hsize), PIL.Image.ANTIALIAS)
img.save('./samples_content/fallen_star_resize.jpg')
```

```python
#vectorize image
pic1_n = pic1.reshape(pic1.shape[0]*pic1.shape[1], pic1.shape[2])
pic1_n.shape
```

```python
#apply k-means
kmeans1 = KMeans(n_clusters=5, random_state=0).fit(pic1_n)
pic2show1 = kmeans1.cluster_centers_[kmeans1.labels_]
```

```python
#show clustered picture
cluster_pic1 = pic2show1.reshape(pic1.shape[0], pic1.shape[1], pic1.shape[2])
plt.imshow(cluster_pic1)
```
